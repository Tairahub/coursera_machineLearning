```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def compute_cost(x, y, w, b):
    m, n = x.shape # m: number of column, n: number of row
    cost = 0
    for i in range(m):
        cost += (-1 * y[i] * np.log(sigmoid(np.dot(x[i], w) + b))) - (1 - y[i]) * (np.log(1 - sigmoid(np.dot(x[i], w) + b)))
    cost /= m
    return cost
```

```python
def compute_gradient(x, y, w, b):
    m = x.shape[0] # m: number of column
    dj_dw = 0
    dj_db = 0
    for i in range(m):
        dj_dw += (sigmoid(np.dot(x[i], w) + b) - y[i]) * x[i]
        dj_db += sigmoid(np.dot(x[i], w) + b) - y[i]
    dj_dw, dj_db = dj_dw / m, dj_db / m
    return dj_db, dj_dw
```

```python
def gradient_descent(x, y, w_init, b_init, alpha, num_iters, lambda_):
    m = x.shape[0] # m: number of column
    w = w_init
    b = b_init
    cost = []
    for i in range(num_iters):
        dj_db, dj_dw = compute_gradient(x, y, w, b)
        b -= alpha * dj_db
        w -= alpha * dj_dw
        cost.append(compute_cost(x, y, w, b))
    return w, b, cost
```

```python
np.random.seed(1)
w_init = 0.01 * (np.random.rand(2) - 0.5)
b_init = -8
iterations = 10000
alpha = 0.001
lambda_ = 0

w, b, cost_history = gradient_descent(x_train, y_train, w_init, b_init, alpha, iterations, lambda_)

print(f"w: {w}, b: {b}")
plt.plot(cost_history)
```

```python
def predict_logistic(x, w, b):
    m = x.shape[0] # m: number of column
    result = np.zeros(m)
    for i in range(m):
        result[i] = sigmoid(np.dot(x[i], w) + b)
        if result[i] >= 0.5:
            result[i] = 1
        else:
            result[i] = 0
    return result
```

```python
def compute_cost_reg(x, y, w, b, lambda_ = 1):
    m, n = x.shape # m: number of column, n: number of row
    cost_without_reg = 0
    for i in range(m):
        cost_without_reg += (-1 * y[i] * np.log(sigmoid(np.dot(x[i], w) + b))) - (1 - y[i]) * (np.log(1 - sigmoid(np.dot(x[i], w) + b)))
    cost_without_reg /= m

    reg_cost = 0
    for i in range(n):
        reg_cost += w[i] ** 2
    reg_cost = reg_cost * lambda_ / (2 * m)

    return cost_without_reg + reg_cost
```

```python
def compute_gradient_reg(x, y, w, b, lambda_=1):
    m = x.shape[0] # m: number of column
    dj_dw = 0
    dj_db = 0
    for i in range(m):
        dj_dw += (sigmoid(np.dot(x[i], w) + b) - y[i]) * x[i]
        dj_db += sigmoid(np.dot(x[i], w) + b) - y[i]
    dj_dw, dj_db = dj_dw / m, dj_db / m
    dj_dw = dj_dw + (lambda_ * w / m)
    return dj_dw, dj_db
```

```python
def gradient_descent_reg(x, y, w_init, b_init, alpha, num_iters, lambda_):
    m = x.shape[0] # m: number of column
    w = w_init
    b = b_init
    cost = []
    for i in range(num_iters):
        dj_dw, dj_db = compute_gradient_reg(x, y, w, b)
        b -= alpha * dj_db
        w -= alpha * dj_dw
        cost.append(compute_cost(x, y, w, b))
    return w, b, cost
```
